Smart Home Device Simulator

A Command-Line Interface (CLI) application written in Python that simulates a basic smart home automation environment. This system models common household smart devices using Object-Oriented Programming (OOP) core pillars like **Inheritance**, **Encapsulation**, and **Polymorphism**.

Project Task Summary

The objective of this project is to construct a modular Python program capable of managing various smart devices.
SmartDevice` (Base Class):** Tracks universal smart hardware elements such as device names, identification strings, and binary power configurations (`True`/`False`).
`temperatureSensor` (Child Class):** Extends the base capabilities to monitor ambient room conditions and handle reading triggers.
`SmartLight` (Child Class):** Adds structural parameters managing hardware brightness thresholds scaled from 0% to 100%.
`SecurityCamera` (Child Class):** Handles operational recording state structures dependent directly on power conditions.

A centralized runtime loop provides interacting capability with these simulated elements, executing logic checks (e.g., preventing light dimming or video recording if the target device is powered down).

Features Implemented
Encapsulation:** Device states and structural details are protected via private attributes (`__device_id`, `__brightness`, etc.) and are only alterable via designated getter/setter methods.
Method Overriding:** Custom `.display_info()` routines dynamically stitch general metrics together alongside child-specific tracking attributes.
State Dependency Management:** Preventative logic handling ensures operations fail gracefully if elements are turned off.

```python
class SmartDevice:
    def __init__(self,name,device_id):
        self.__device_id = device_id
        self.name =name
        self.__power_status = False

    def get_device_id(self):
        return self.__device_id
    def get_power_status(self):
        return self.__power_status
    def set_power_status(self, power_status):
        self.__power_status = power_status
    def turn_on(self):
        self.__power_status = True
        print(self.name,"has been turned on successfully")
    def turn_off(self):
        self.__power_status = False
        print(self.name,"has been turned off successfully")
    def display_info(self):
        print("Device name is", self.name)
        print("Device ID is", self.__device_id)
        print("Power is", self.__power_status)



class temperatureSensor(SmartDevice):
    def __init__(self,device_id,name,temperature):
        super().__init__(name,device_id)
        self.temperature = temperature
    def read_temperature(self):
        a=input("would you like to read the temperature?")
        if a=="yes":
            print(self.name,"is the temperature of the room")
        else:
            pass
    def display_info(self):
        super().display_info()
        print("The temperature is", self.temperature)


class SecurityCamera(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.__recording_status = False

    def get_recording_status(self):
        return self.__recording_status

    def start_recording(self):
        self.recording_status = True
        if self.is_active():
            self.__recording_status = True
            print(self.name + " has started recording video.")

    def stop_recording(self):
        self.__recording_status = False
        print(self.name + " has stopped recording video.")

    def display_info(self):
        super().display_info()
        status_text = "Recording" if self.__recording_status else "Idle"
        print("Camera State: " + status_text)


class SmartLight(SmartDevice):
    """Child Class representing a dimmable smart light."""

    def __init__(self, name, device_id, brightness=50):
        super().__init__(name, device_id)
        if brightness < 0 or brightness > 100:
            raise ValueError("Brightness must be between 0 and 100.")
        self.__brightness = brightness  # Private attribute

    def get_brightness(self):
        return self.__brightness

    def set_brightness(self, amount):
        if amount < 0 or amount > 100:
            print("Error: Brightness must be between 0 and 100.")
        else:
            self.__brightness = amount

    def increase_brightness(self):
        # Uses parent check to skip adjustment if the light is off
        if self.is_active():
            new_brightness = self.__brightness + 10
            self.__brightness = min(new_brightness, 100)
            print(self.name + " brightness increased to " + str(self.__brightness) + "%")

    def decrease_brightness(self):
        # Uses parent check to skip adjustment if the light is off
        if self.is_active():
            new_brightness = self.__brightness - 10
            self.__brightness = max(new_brightness, 0)
            print(self.name + " brightness decreased to " + str(self.__brightness) + "%")

    def display_info(self):
        super().display_info()
        print("Brightness:   " + str(self.__brightness) + "%")


def main():
    sensor = temperatureSensor("Living Room Temp Sensor", "TS-101", 24.5)
    light = SmartLight("Kitchen Light", "SL-202", 70)
    camera = SecurityCamera("Front Door Camera", "SC-303")

    while True:
        print("\n=== SMART DEVICE MENU ===")
        print("1. Display Device Information")
        print("2. Turn Device On")
        print("3. Turn Device Off")
        print("4. Read Temperature")
        print("5. Adjust Brightness")
        print("6. Recording Controls")
        print("7. Exit")

        choice = input("Enter choice (1-7): ")

        if choice == "1":
            print("\n--- DEVICES STATUS ---")
            sensor.display_info()
            print("----------------------")
            light.display_info()
            print("----------------------")
            camera.display_info()
            print("----------------------")

        elif choice == "2":
            print("\nWhich device to turn ON?")
            print("1. Temperature Sensor")
            print("2. Smart Light")
            print("3. Security Camera")
            dev_choice = input("Choose (1-3): ")
            if dev_choice == "1":
                sensor.turn_on()
            elif dev_choice == "2":
                light.turn_on()
            elif dev_choice == "3":
                camera.turn_on()
            else:
                print("Invalid choice.")

        elif choice == "3":
            print("\nWhich device to turn OFF?")
            print("1. Temperature Sensor")
            print("2. Smart Light")
            print("3. Security Camera")
            dev_choice = input("Choose (1-3): ")
            if dev_choice == "1":
                sensor.turn_off()
            elif dev_choice == "2":
                light.turn_off()
            elif dev_choice == "3":
                camera.turn_off()
            else:
                print("Invalid choice.")

        elif choice == "4":
            print("\n--- Temperature Reading ---")
            sensor.read_temperature()

        elif choice == "5":
            print("\n--- Adjust Brightness ---")
            print("1. Increase Brightness (+10%)")
            print("2. Decrease Brightness (-10%)")
            adjust_choice = input("Choose (1-2): ")
            if adjust_choice == "1":
                light.increase_brightness()
            elif adjust_choice == "2":
                light.decrease_brightness()
            else:
                print("Invalid choice.")

        elif choice == "6":
            print("\n--- Recording Controls ---")
            print("1. Start Recording")
            print("2. Stop Recording")
            rec_choice = input("Choose (1-2): ")
            if rec_choice == "1":
                camera.start_recording()
            elif rec_choice == "2":
                camera.stop_recording()
            else:
                print("Invalid choice.")

        elif choice == "7":
            print("\nExiting program. Goodbye!")
            break
        else:
            print("Invalid choice! Please select 1 to 7.")


if __name__ == "__main__":
    main()





```
